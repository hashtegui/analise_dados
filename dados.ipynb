# Analisando Vendas de Clientes

```python
import pandas as pd


df = pd.read_excel("Dados/dados.xls")
dados = df
```

## Somando compras de cliente pela categoria

```python
# SEPARANDO DADOS IMPORTANTES
dados = dados[['CODCLI', 'NOME_CLIENTE', 'EMAIL_CLIENTE', 'CODCATEGORIA','CATEGORIA']]

# FAZENDO A SOMA DA CATEGORIA POR COMPRA DE CLIENTE

dados = dados.groupby(['CODCLI', 'NOME_CLIENTE', 'EMAIL_CLIENTE', 'CATEGORIA']).count()
```

```python
dados.to_excel('Dados/planilha_nova.xls')
```
